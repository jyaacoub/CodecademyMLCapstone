# Question:
Can I predict age based on the frequency of certain words in user written essays?

## Regressors used:
* Multilinear regressor
* KNN regressor

## Imports needed:

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

from sklearn.feature_extraction.text import CountVectorizer
```

# Getting the Data:
Github only allows files < 100mb so I can only upload a truncated version of the file.

```python
# Original file: C:\Users\jyaac\Documents\codecademy_projects\Machine Learning\capstone_starter\profiles.csv
# Truncated file: profiles_trimmed.csv
df = pd.read_csv(r'..\capstone_starter\profiles.csv')
```

### Cleaning the data, and creating `all_essays` column:
Replacing na values with empty essays, and creating an all_essays column to hold all the essays.

```python
# There are 10 possible short answer essays that a user can write (named essay0, essay1, ..., essay9)
essays = ['essay'+str(i) for i in range(10)]
temp = df[essays].replace(np.nan, '', regex=True) # nans are just empty essays

# the passed in value is the series of objects across a data frame index (row).
df['all_essays'] = temp[essays].apply(lambda x: ' '.join(x), axis=1)
```

### Creating CountVectorizer:
To transform the data so that it can be run through the regressors.
CountVectorizer doesn't consider single letter words as words so the regex below changes that (from https://www.w3schools.com/python/python_regex.asp).
    
* `ngram_range` is the frequency count use (monogram, bigram, or trigram)
    
     example: (1,1) is only monogram and (1,2) is both monogram and bigram, and so on.

```python
# Creating the vocab to transform the data into count vectors
vocab = np.array(['i', 'me', 'like', 'lit', 'literally'])
pattern = u'(?u)\\b\\w+\\b'     # default has an extra w: u'(?u)\\b\\w\\w+\\b'

counter = CountVectorizer(token_pattern=pattern)    # default ngram_range is (1,1), and it also lowercases all inputs
counter.fit(vocab)  # It will only count frequencies of these words

feat_names = counter.get_feature_names() # order of vocab list is alphabetically sorted
print('VOCAB:', feat_names)
```

```python
features = counter.transform(df['all_essays'])
labels = df['age']
print(features.shape)
print(labels.shape)
```

### Exploring the data:
Checking the frequencies of the word counts.

```python
print(features.shape)
total_counts = [0]* features.shape[1]

for individual in features.toarray():
    for j in range(len(total_counts)):
        total_counts[j] += individual[j]
```

```python
print(total_counts)
print(feat_names)

plt.hist(feat_names, bins=len(feat_names), weights=total_counts)
plt.show()
```

Looking for any relationships between some words and age. This doesn't really give me much information because the trend might just be because there are way more younger people using dating apps.

```python
like_s = [x[1] for x in features.toarray()]
print(len(like_s))
print(len(labels))

plt.scatter(labels, like_s, alpha= 0.1)
plt.xlabel('age')
plt.ylabel('Frequency of i\'s')
plt.xlim(17,80)
# plt.ylim(0,300)

plt.show()
```

### Splitting the data
Creating training and testing sets using our created `all_essays` column created earlier as feature data and age as labels.

```python
train_data, test_data, train_labels, test_labels = train_test_split(features, labels, train_size=0.8, random_state=23)
print(train_data.shape, test_data.shape)
print(train_labels.shape, test_labels.shape)
```

# Training and testing the models
## Multilinear Regressor


### Creating the model:
Fitting it to the training data and determining its coefficient of determination (R^2).

```python
linear_model = LinearRegression(normalize=True)
linear_model.fit(train_data, train_labels)
print(linear_model.score(test_data, test_labels))
```

### Further evaluations
Ploting predicted values on scatter plot of age vs frequency of the word 'like'.

As the number of likes increases what happens to the predicted age?

```python
from scipy.sparse import csr_matrix

# must be a n x 5 matrix:
print(features.shape)
print(type(features))

# the frequency of the word 'like' is stored in the second index position:
print(feat_names)
```

```python
n = 50

# all other values will be set to 0
rows = np.array([[0,x,0,0,0] for x in range(0, n)])


like_matrix = csr_matrix(rows).toarray()
print(like_matrix[:5])

prediction = linear_model.predict(like_matrix)
print(prediction[:5])
```

```python
# Isolating the likes:
num_like = features[:,1].toarray() 
print(num_like[:3])

# scatter plot of frequency of 'like' vs. age
plt.scatter(labels, num_like, alpha=0.1)
plt.xlabel('age')
plt.ylabel('Frequency of like\'s')
plt.xlim(17,80)

# ploting the prediction like:
likes_range = list(range(n)) 
plt.plot(prediction, likes_range, color='red')
plt.show()
```

From this we don't gain much information other than that the frequency of likes doesnt change much when we set all other values to 0 and are only incrementing the frequency of likes

## K-Nearest Neighbor Regressor:

### Training and testing on default settings:

```python
model = KNeighborsRegressor()
model.fit(train_data, train_labels)
print(model.score(test_data, test_labels))
```

### Ploting `n_neighbors` vs. coeff of determination:

**Getting the data:**

```python
k_list = list(range(1, 10))
results = []
for k in k_list:
    print(k, end=', ')
    model = KNeighborsRegressor(n_neighbors=k,
                                weights='distance')     # distance indicates that for things further away they have
                                                        # a greater impact on the prediction (WEIGHTED)
    model.fit(train_data, train_labels)
    score = model.score(test_data, test_labels)
    results.append(score)
    print(score)
```

**Ploting the data:**

```python
plt.plot(k_list, results)
plt.xlabel('n_neighbors')
plt.ylabel('coeff of determination (R^2)')
plt.show()
```
